The maximum of the four is: 6

```python
# Q2
import sys

# calculating average and max with in-build func
def average(nums):
  t = 0
  for i in nums:
    t+=i
  return t/len(nums)

def get_max(nums):
  high = -sys.maxsize

  for i in nums:
    if i > high:
      high = i

  return high


try:
  num_count = int(input('Enter set of 3 or 4 integers: '))
  num_eng = ''
  # checking if user entered 3 or 4
  if num_count in [3, 4]:
    num1 = int(input('Enter first number: '))
    num2 = int(input('Enter second number: '))
    num3 = int(input('Enter third number: '))
    nums = [num1, num2, num3]
    num_eng = 'three'
    if num_count == 4:
      num4 = int(input('Enter fourth number: '))
      nums.append(num4)
      num_eng = 'four'
    
    print('The average of', nums, 'is:', average(nums))
    print('The maximum of the', num_eng, 'is:', get_max(nums))
  else:
    print('Enter int 3 or 4 only')

except ValueError:
  print('Error: You did not enter an integer')
```

```python
# Q3

try:
  num = int(input('Enter positive integer: '))

  if num > 0:
    # checking edge case
    if num == 1:
      print('It is neither prime nor composite')
    else:
      divisible = []
      
      for i in range(1, num+1):
        # checking if divisible
        if num % i == 0:
          divisible.append(i)   # storing num which are divisors

      # using ternary operator to check if prime or composite
      print('It is a prime') if divisible == [1, num] else print('It is a composite')
  else:
    raise ValueError()

except ValueError:
  print('Error: You did not enter a positive integer')
```

```python
# Q4

# returns the simple interest amount
def SimpleInterest(principle, interest, years):
  return (principle*interest*years)/100

try:
  principle = float(input('Enter Principle: '))
  interest = float(input('Enter interest (0 to 100 as a float): '))
  try:
    years = int(input('Enter years: '))
    print(SimpleInterest(principle, interest, years))   # making func call
  except ValueError:
    print('Error: You did not enter an int')
except ValueError:
  print('Error: You did not enter a float')
```

```python
# Q5

prd, cost = [], []

try:
  while True:
    print('Hit enter to stop adding products to cart')
    product = input('Enter Product Name: ')
    # user can break the loop by hitting enter
    if product == '':         
      break
    price = float(input('Enter price: '))

    if price <= 0:
      print('price should be more than $0')
      break
    else:
      # storing product names and price
      prd.append(product)
      cost.append(price)
  
  print(f'Total Bill = ${sum(cost):,.2f}')    # used to display bill till 2 decimals
  
  # displaying all items in cart
  print('|Product Name | Price |')
  for p,c in list(zip(prd, cost)):
    print(p, '  ', c)

except ValueError:
  print('Error: You did not enter a float for price')
```
